# ImSpire: Color scheme inspirations from images

This is a tool for designers to come up with an new color schemes from Images.
Coded in Python.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from PIL import Image
```

In this example, we will be using the following image of a gazelle.

```python
picture = "gazellepic.jpg"
im = Image.open(picture)
im
```

## Resizing

We resize the image, reducing the number of pixels. This helps in better plotting time.

```python
resize_factor = 0.05
resize_params = tuple([int(x*resize_factor) for x in im.size])
im_resized = im.resize(resize_params,Image.ANTIALIAS)
im_values = np.array(im_resized)

im_resized
```

## Getting RGB values as seperate arrays

RGB values of all pixels stored as separate r, g and b arrays.

```python
r = im_values[:,:,0].flatten()
g = im_values[:,:,1].flatten()
b = im_values[:,:,2].flatten()

colors = np.array([[r,g,b] for r,g,b in zip(r,g,b)])
```

## Plotting 3D color map

We plot 3D map, axes are R,G and B values. Each point in scatter plot corresponds to one pixel.

```python
%matplotlib notebook

colors = np.array([[r,g,b, 0.2*255] for r,g,b in zip(r,g,b)])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(r,g,b, c= colors/255.0)
plt.show()
```
